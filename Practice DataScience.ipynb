# CSV FILE DOWNLOAD, CLEANING, VISUALISING 

```python
#CSV DOWNLOAD
df = pd.read_csv('myfin.csv')
df
```

```python
#CLEANING THE DATAFRAME
dn = df.values
dn
```

```python
#CLEAN THE FOODING
f_l = []
i = 0
while i <= 6:
    f_l.append(dn[i,2])
    i+=1
f_l[3]=0          
f_l[5]=0                 #FILLING NAN WITH 0

mf_l = round(st.mean(f_l), 2)
mf_l
```

```python
#CLEAN THE TRAVELLING LIST
t_l = []
i = 0
while i <= 6:
    t_l.append(dn[i,3])
    i+=1
t_l[3] = 0
t_l[4] = 0
mt_l = round(st.mean(t_l), 2)
mt_l
```

```python
#CLEAN THE HEALTH LIST
h_l = []
i = 0
while i <= 6:
    h_l.append(dn[i,4])
    i+=1
h_l[4] = 0
h_l[5] = 0
mh_l = round(st.mean(h_l),2)
mh_l
```

```python
#CLEAN DATA INPUT
df.loc[3, 'F=Fooding(I/4)'] = mf_l
df.loc[5, 'F=Fooding(I/4)'] = mf_l
df.loc[3, 'T=Travelling(I/5)']= mt_l
df.loc[4, 'T=Travelling(I/5)']= mt_l
df.loc[4, 'H=Health(I/3)'] = mh_l
df.loc[5, 'H=Health(I/3)'] = mh_l
#NEW DATAFRAME
df
```

```python
#FINDING THE SAVING AT 6TH INDEX
s = (df.loc[6,'Income(I)']-(df.loc[6,'F=Fooding(I/4)'] + df.loc[6,'T=Travelling(I/5)'] + df.loc[6,'H=Health(I/3)']))
df.loc[6,'S=Savings(I-(F+T+H))'] = s
df
```

# WEATHER PLOT FOR 7 DAYS USING PANDAS

```python
#IMPORTING THE DATE RANGE FOR TODAY TILL 7 DAYS
ind = pd.date_range('today', periods=7)
#WEATHER RANDOM VARIABLES
import random
w = np.random.randn(7)
w = (abs(w) * 10)
#CLEARING THE ARRAY USING MAP
def myfunc(r):
    if r > 10:
        r= r+15
    else:
        r= r+23
    return r

n = list(map(myfunc,w))
m = []
#RANDOMISING THE PLACE OF VARIABLES
for _ in range(0,7):
    x = random.randint(0,6)
    m.append(n[x])
# CREATING THE PANDAS SERIES
s1 = pd.Series(m, index= ind)
# PLOTTING THE WEATHER REPORT
s1.cumsum()
s1.plot()
```
