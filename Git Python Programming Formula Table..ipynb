```python
import numpy as np, pandas as pd
from matplotlib import pyplot as plt
```

```python
# Our objective is to create a table which will show the different function of a variable
x = np.arange(1,21)
# First function 
f_1 = x*x
# Second function
f_2 = 1/(x*x)
# Third function
f_3 = (np.power((np.e),x)) + np.power((np.e),(-x))
# Forth function
f_4 = (np.power(x,3) + 2*np.power(x,2) + 2*x + 2)
# Fifth function(Derivative of f_4)
f_5 = (3*np.power(x,2) + 4*np.power(x,1) + 2)
# Sixth function
f_6 = np.log(x)
```

```python
# Lets create a pandas df
df = pd.DataFrame({'Variable(x)':x, 'f(x)=x^2':f_1, 'f(x)=1/x^2':f_2, 'f(x)e^x + e^-x':f_3, 
                   'f(x)=x^3+2x^2+2x+2': f_4, 'f(x)=3x^2+4x+2': f_5, 'f(x)= log(x)': f_6})
df
```

```python
from matplotlib import pylab as plb
```

```python
fig, axes = plt.subplots(1,6, figsize=(20,9))
axes[0].set_title('Square function')
axes[0].plot(df['Variable(x)'], df['f(x)=x^2'], lw=2)

axes[1].set_title('Inverse Square function')
axes[1].plot(df['Variable(x)'], df['f(x)=1/x^2'], 'red', lw=2)

axes[2].set_title('Hyperbolic Sine')
axes[2].plot(df['Variable(x)'], df['f(x)e^x + e^-x'], 'green', lw=2)

axes[3].set_title('Cubic-Polynomial')
axes[3].plot(df['Variable(x)'], df['f(x)=x^3+2x^2+2x+2'], 'purple', lw=2)

axes[4].set_title('Derivative of the previous')
axes[4].plot(df['Variable(x)'], df['f(x)=3x^2+4x+2'], 'orange', lw=3)

axes[5].set_title('Logarithm')
axes[5].plot(df['Variable(x)'], df['f(x)= log(x)'], 'brown', lw=2)
```

```python
import random
def poll():
    polled= []
    for _ in range(0,6):
       
        r = random.randint(1,100)
        r1 = random.randint(1,(100-r))
        r2 = random.randint(1,(100-(r+r1)))
        r3 = random.randint(1,(100-(r+r1+r2)))
        r4 = random.randint(1,(100-(r+r1+r2+r3)))
        r_f = [r,r1,r2,r3,r4]
        polled.append(r_f)
        r_f = [0,0,0,0,0]
        
        continue
    return polled
p = poll()
p
```

```python
import numpy as np, pandas as pd
```

```python
data = pd.read_csv('Food ML.csv')
indexs = data['Resturants']
data_1 = data.drop('Resturants', axis=1)
n= data_1.values
data_2 = pd.DataFrame({'Mutton':n[:,0], 'Ghee':n[:,1], 'Rice':n[:,2], 
                      'Aroma':n[:,3], 'Aloo':n[:,4], 'Egg':n[:,5], 'Quality':n[:,6]}, index= indexs)
data_2
```

```python
# Categorising the quality in Good or Bad(Statistical Categorisation)
bins = (3, 7.5, 9)
group_name = ['Bad', 'Good']
data_2['Quality'] = pd.cut(data_2['Quality'], bins= bins, labels= group_name)
data_2['Quality'].unique()
```

```python
data_2
```

```python
from sklearn.preprocessing import LabelEncoder
label_quality = LabelEncoder()
data_2['Quality']= label_quality.fit_transform(data_2['Quality'])
data_2
```

```python
import seaborn as sns
sns.countplot(data_2['Quality'])
```

```python
x= data_2.drop('Quality', axis=1)
y= data_2['Quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.5, random_state= 42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
pred_1 = rfc.predict(x_test)
```

```python
pred_1
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, pred_1))
```

```python
print(confusion_matrix(y_test, pred_1))
```
